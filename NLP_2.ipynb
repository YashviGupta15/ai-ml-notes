# NLP Preprocessing Pipeline

Raw Text → Tokenization → Lowercase → Stopword Removal → Stemming → Lemmatization

---

# Step 1 — Tokenization

Tokenization converts sentence into words

Example:

"I love NLP"

↓

["I", "love", "NLP"]

Mathematically:

Sentence:

$S = [w_1, w_2, w_3, ..., w_n]$

---

# Step 2 — Lower Casing

"I Love NLP"

↓

Why?

"I" and "i" should represent same word

---

# Step 3 — Stop Words Removal

Common words:

- is
- the

- of

Example:

"This is a machine learning model"

↓

["machine", "learning", "model"]

---

# Step 4 — Stemming

Reducing word to root

Example:

But:

Sometimes incorrect

Example:

---

# Step 5 — Lemmatization

Better version of stemming

running → run

---

# Final Pipeline Example

# Why This Matters

After preprocessing:

- Vocabulary size reduces
- Model learns better

- Performance improves

---

# Bridge to Next Step

Now we have clean tokens:

["learn", "machine", "learn"]

Next step:

Convert words into numbers

```python
# Text Preprocessing Example

import nltk
import re

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.corpus import wordnet

# Downloads (Run once)
#nltk.download('punkt')
#nltk.download('stopwords')
#nltk.download('wordnet')
#nltk.download('averaged_perceptron_tagger')

sentence = "I am learning machine learning and loving it"

# Lowercase
sentence = sentence.lower()

# Tokenization
tokens = nltk.word_tokenize(sentence)

# Remove punctuation
tokens = [re.sub(r'[^\w\s]', '', word) for word in tokens]

# Remove empty strings
tokens = [word for word in tokens if word != ""]

# Remove stopwords
stop_words = set(stopwords.words('english'))
tokens = [word for word in tokens if word not in stop_words]

# Stemming
stemmer = PorterStemmer()
stemmed = [stemmer.stem(word) for word in tokens]

# Function for Lemmatization with POS tagging

def get_wordnet_pos(word):
    tag = pos_tag([word])[0][1][0].upper()
    tag_dict = {
        "J": wordnet.ADJ,
        "N": wordnet.NOUN,
        "V": wordnet.VERB,
        "R": wordnet.ADV
    }
    return tag_dict.get(tag, wordnet.NOUN)

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in tokens]

print("Original Sentence:")
print("I am learning machine learning and loving it")

print("\nTokens:")
print(tokens)

print("\nStemmed:")
print(stemmed)

print("\nLemmatized:")
print(lemmatized)
```

Assign index to each word:

This is called:

# Word Dictionary Mapping

---

# Convert Sentence to Numbers

---

# Mathematical Representation

Vocabulary:

$V = \{w_1, w_2, w_3, ... , w_n\}$

Index mapping:

$f(w_i) = i$

Sentence:

$S = [w_1, w_2, w_3]$

Numeric Representation:

$S = [1,2,3]$

---

# Why This is Important

Now we converted:

Text → Numbers

But there is still a problem:

[0,1,2] does NOT capture meaning

Example:

king → 1  
queen → 2  

Does this mean queen > king?

No.

Numbers introduce **false relationships**

This leads to:

One Hot Encoding (Next Section)

```python
# Vocabulary Creation Example

sentences = [
    "i love nlp",
    "i love deep learning",
    "deep learning is powerful"
]

# Tokenization
tokens = [sentence.split() for sentence in sentences]

# Create vocabulary
vocab = set()

for sentence in tokens:
    for word in sentence:
        vocab.add(word)

# Convert to dictionary
word2index = {word: i for i, word in enumerate(vocab)}

print("Vocabulary:\n")
print(word2index)

# Convert sentence to numeric

example = "i love nlp".split()
numeric = [word2index[word] for word in example]

print("\nSentence:")
print(example)

print("\nNumeric Representation:")
print(numeric)
```

Observation:

- Words converted to numbers
- Vocabulary created
- Sentence converted to numeric sequence

But numbers do not represent meaning

Example:

king → 1  
queen → 2  

Model assumes numeric relationship

This is incorrect

We need better representation

# One Hot Encoding

In Vocabulary Mapping:

king → 0  
queen → 1  
man → 2  
woman → 3  

Sentence:

"I love NLP"

↓

[0,1,2]

But numeric indexing introduces **false relationships**

queen (1) > king (0)

This is incorrect.

---

# Solution — One Hot Encoding

Instead of using numbers directly,  
we represent each word as **vector**

Example Vocabulary:

{king, queen, man, woman}

Vector Representation:

king  → [1,0,0,0]  
queen → [0,1,0,0]  
man   → [0,0,1,0]  
woman → [0,0,0,1]

Each word has **unique vector**

---

# Properties of One Hot Encoding

- Only one position is 1
- All others are 0
- Vector size = Vocabulary size

---

# Mathematical Representation

Vocabulary size = V

Each word represented as:

One Hot Vector:

Dimension = V

---

# Example

Vocabulary:

{I, love, NLP}

I → [1,0,0]  
love → [0,1,0]  
NLP → [0,0,1]

Sentence:

"I love NLP"

↓

[[1,0,0],  
 [0,1,0],  
 [0,0,1]]

---

# Why One Hot Encoding is Better

- No numeric bias
- No false ordering
- Unique representation

But still problem exists:

king → [1,0,0,0]  
queen → [0,1,0,0]

Similarity = 0

But king and queen are related.

One Hot Encoding cannot capture similarity

This leads to:

Word Embeddings

```python
# One Hot Encoding Example

import numpy as np

sentences = [
    "i love nlp",
    "i love deep learning"
]

# Build vocabulary
vocab = set()

for sentence in sentences:
    for word in sentence.split():
        vocab.add(word)

vocab = list(vocab)

# Create mapping
word2index = {word: i for i, word in enumerate(vocab)}

# One hot encoding

def one_hot(word):
    vector = np.zeros(len(vocab))
    vector[word2index[word]] = 1
    return vector

# Example

word = "deep"

print("Vocabulary:\n", vocab)
print("\nOne Hot Encoding for 'deep':\n")
print(one_hot(word))
```

Observation:

- Each word has unique vector
- No numeric relationship
- No ordering problem

But:

All words equally distant

Similarity not captured

Example:

king vs queen → similarity = 0

This is not ideal

We need semantic representation

# Word2Vec — Intuition

In Word Embeddings:

We said:

Words appearing in similar context  
Have similar meaning

Word2Vec learns embeddings using **context**

---

# Core Idea

Given sentence:

"I love deep learning"

Choose context window = 2

Context:

I → love  
love → I, deep  
deep → love, learning  
learning → deep

---

# Context Window

Sentence:

"I love deep learning"

Window Size = 1

Training pairs:

(I → love)

(love → I)

(love → deep)

(deep → love)

(deep → learning)

(learning → deep)

---

# Two Word2Vec Architectures

1. Skip-Gram  
2. CBOW (Continuous Bag of Words)

---

# Skip-Gram

Given word → Predict context

Example:

"I love deep learning"

Input:

love

Output:

I, deep

---

# CBOW

Given context → Predict word

Input:

I, deep

Output:

love

---

# Why This Works

Words with similar context  
Learn similar embeddings

Example:

"I drink coffee"

"I drink tea"

coffee and tea appear in similar context

Thus embeddings become similar

```python
# Word2Vec Training Pair Example

sentence = "i love deep learning".split()
print(len(sentence))

window_size = 1

pairs = []

for i, word in enumerate(sentence):

    for j in range(-window_size, window_size + 1):

        if j != 0 and i + j >= 0 and i + j < len(sentence):
            pairs.append((word, sentence[i + j]))

print("Training Pairs:\n")

for pair in pairs:
    print(pair)
```

Observation:

Training pairs created automatically

Model learns:

Words with similar context  
Have similar embeddings

This is how Word2Vec learns meaning

# Co-Occurrence Matrix

Word2Vec learns using **local context**

But what if we use:

Global context?

This leads to:

# Co-Occurrence Matrix

---

# Core Idea

Words that appear together  
Have related meaning

Example Sentences:

"I love deep learning"  
"I love machine learning"  
"deep learning is powerful"

---

# Vocabulary

{I, love, deep, learning, machine, is, powerful}

---

# Co-Occurrence Concept

Count how often words appear together

Example:

"I love deep learning"

Window size = 1

Pairs:

(I, love)  
(love, deep)  
(deep, learning)

---

# Co-Occurrence Matrix

| Word | I | love | deep | learning |
|------|---|------|------|----------|
| I | 0 | 1 | 0 | 0 |
| love | 1 | 0 | 1 | 0 |
| deep | 0 | 1 | 0 | 1 |
| learning | 0 | 0 | 1 | 0 |

---

# Interpretation

love and deep appear together  
deep and learning appear together

This creates **relationship between words**

---

# Mathematical Representation

Co-Occurrence Matrix:

$X_{ij}$ = Number of times word i appears with word j

---

# Why This is Useful

Words appearing together frequently  
Have similar meaning

Example:

coffee appears with:

drink  
cup  
morning

tea appears with:

drink  
cup  
morning

Thus:

coffee ≈ tea

```python
# Co-Occurrence Matrix Example

import numpy as np
import pandas as pd

sentences = [
    "i love deep learning",
    "i love machine learning",
    "deep learning is powerful"
]

# Build vocabulary
vocab = list(set(" ".join(sentences).split()))

word2index = {word:i for i,word in enumerate(vocab)}

# Initialize matrix
co_matrix = np.zeros((len(vocab), len(vocab)))

window_size = 1

for sentence in sentences:

    words = sentence.split()

    for i, word in enumerate(words):

        for j in range(-window_size, window_size+1):

            if j != 0 and i+j >= 0 and i+j < len(words):

                neighbor = words[i+j]

                co_matrix[word2index[word]][word2index[neighbor]] += 1

# Convert to DataFrame
co_df = pd.DataFrame(co_matrix, index=vocab, columns=vocab)

print("Co-Occurrence Matrix:\n")
print(co_df)
```

Observation:

Words appearing together frequently  
Have higher co-occurrence values

Example:

deep — learning → High value

This indicates:

Words are related
