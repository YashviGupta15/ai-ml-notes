# Recurrent Neural Network (RNN) Architecture

So far we understand:

- Sequential data has order
- Previous information matters
- Model needs **memory**

RNN introduces **Hidden State** as memory.

---

# Basic RNN Architecture

At each time step, RNN performs:

Input → Hidden State → Output

But unlike ANN:

Hidden state is **carried forward in time**

This makes RNN **recurrent**

---

# Single Time Step RNN

At time step **t**:

Input:
- Current input → $x_t$
- Previous memory → $h_{t-1}$

Output:
- New memory → $h_t$
- Output → $y_t$

---

# Mathematical Formulation

Hidden State Update

$$
h_t = \tanh(W_x x_t + W_h h_{t-1} + b)
$$

Output Generation

$$
y_t = W_y h_t + b_y
$$

---

# What is Happening?

At each time step:

1. Take new input
2. Combine with previous memory
3. Produce updated memory
4. Generate output

This is the **core mechanism of RNN**

---

# Tensor Shapes (Important)

Assume:

Batch size = $B$  
Input size = $D$  
Hidden size = $H$  
Output size = $O$

Input:

$$
x_t \rightarrow (B, D)
$$

Previous Hidden State:

$$
h_{t-1} \rightarrow (B, H)
$$

Weights:

$$
W_x \rightarrow (H, D)
$$

$$
W_h \rightarrow (H, H)
$$

$$
W_y \rightarrow (O, H)
$$

Outputs:

$$
h_t \rightarrow (B, H)
$$

$$
y_t \rightarrow (B, O)
$$

---

# Why This is Powerful

Because:

- Memory flows across time
- Model learns sequential patterns
- Past influences future

This is what makes RNN suitable for:

- Language
- Time series
- Speech
- Sequential prediction

```python
import torch

# Batch size
B = 2

# Dimensions
D = 5   # input size
H = 4   # hidden size
O = 3   # output size

# Input
x_t = torch.randn(B, D)

# Previous hidden state
h_prev = torch.randn(B, H)

# Weights
W_x = torch.randn(H, D)
W_h = torch.randn(H, H)
W_y = torch.randn(O, H)

# Bias
b = torch.randn(H)
b_y = torch.randn(O)

# Forward pass
h_t = torch.tanh(x_t @ W_x.T + h_prev @ W_h.T + b)
y_t = h_t @ W_y.T + b_y

print("Input shape:", x_t.shape)
print("Previous hidden shape:", h_prev.shape)
print("New hidden shape:", h_t.shape)
print("Output shape:", y_t.shape)
```

# Unrolled RNN

So far we saw:

Single RNN cell at one time step

But sequential data contains **multiple time steps**

Example:

Sentence:

"I love deep learning"

This becomes:

$x_1$ → "I"  
$x_2$ → "love"  
$x_3$ → "deep"  
$x_4$ → "learning"

---

# RNN Across Time

Instead of using one RNN cell, we **repeat the same RNN across time**

Time Step Expansion:

Time 1:

Input → $x_1$  
Hidden → $h_1$

Time 2:

Input → $x_2$  
Hidden → $h_2$

Time 3:

Input → $x_3$  
Hidden → $h_3$

Time 4:

Input → $x_4$  
Hidden → $h_4$

---

# Unrolled RNN Structure

The RNN unfolds across time:

$$
x_1 \rightarrow h_1 \rightarrow x_2 \rightarrow h_2 \rightarrow x_3 \rightarrow h_3 \rightarrow ...
$$

More precisely:

$$
h_1 = f(x_1, h_0)
$$

$$
h_2 = f(x_2, h_1)
$$

$$
h_3 = f(x_3, h_2)
$$

$$
h_t = f(x_t, h_{t-1})
$$

---

# Important Observation

Each time step:

- Takes new input
- Uses previous memory
- Produces new memory

This creates **temporal dependency**

Future depends on past.

---

# Same Weights Across Time

Important property:

The same weights are used at every time step:

$$
W_x, W_h, W_y
$$

are shared across:

- Time step 1
- Time step 2
- Time step 3
- Time step t

This is called:

# Weight Sharing

---

# Why Weight Sharing?

Because:

- Sequence length can vary
- Model must generalize across time
- Fewer parameters required

This makes RNN:

- Efficient
- Scalable
- Generalizable

---

# Why This Matters

Because:

Language:

"not good" vs "very good"

Meaning depends on **previous words**

RNN learns these temporal relationships.

```python
import torch

# Sequence length
T = 4

# Dimensions
B = 2
D = 5
H = 4

# Input sequence
x = torch.randn(T, B, D)

# Initial hidden state
h = torch.zeros(B, H)

# Shared weights
W_x = torch.randn(H, D)
W_h = torch.randn(H, H)
b = torch.randn(H)

print("Unrolled RNN")

for t in range(T):
    h = torch.tanh(x[t] @ W_x.T + h @ W_h.T + b)
    print(f"Time step {t+1} hidden shape:", h.shape)
```

# RNN Forward Pass

Now we understand:

- RNN architecture  
- Unrolled RNN  
- Hidden state flow  

Now let's understand:

# How RNN Computes Forward Pass

At each time step, RNN performs two operations:

1. Hidden State Update  
2. Output Generation  

---

# Hidden State Update

At time step $t$:

Current input → $x_t$  
Previous memory → $h_{t-1}$  

RNN combines both to produce new memory:




# Understanding Each Term

$W_x x_t$  
→ Current input influence

$W_h h_{t-1}$  
→ Past memory influence

$b$  
→ Bias term

$\tanh$  
→ Non-linearity to control values

---

# Output Generation

Once hidden state is computed:

$$
y_t = W_y h_t + b_y
$$

This generates the output at time step $t$

---

# Full Forward Flow

At each time step:

Step 1:

$$
h_t = \tanh(W_x x_t + W_h h_{t-1} + b)
$$

Step 2:

$$
y_t = W_y h_t + b_y
$$

This repeats across time steps.

---

# Tensor Shapes

Assume:

Batch size = $B$  
Input size = $D$  
Hidden size = $H$  
Output size = $O$

Input:

$$
x_t \rightarrow (B, D)
$$

Hidden State:

$$
h_{t-1} \rightarrow (B, H)
$$

Weights:

$$
W_x \rightarrow (H, D)
$$

$$
W_h \rightarrow (H, H)
$$

$$
W_y \rightarrow (O, H)
$$

Outputs:

$$
h_t \rightarrow (B, H)
$$

$$
y_t \rightarrow (B, O)
$$

---

# Why tanh?

Why not linear?

Because linear model cannot learn complex relationships.

$\tanh$:

- Adds non-linearity  
- Controls exploding values  
- Keeps output between:

$$
-1 \leq h_t \leq 1
$$

This stabilizes learning.

---

# Full Sequence Computation

For sequence:

$x_1, x_2, x_3, x_4$

RNN computes:

$$
h_1 = f(x_1, h_0)
$$

$$
h_2 = f(x_2, h_1)
$$

$$
h_3 = f(x_3, h_2)
$$

$$
h_4 = f(x_4, h_3)
$$

Each step depends on previous step.

This is **Sequential Learning**

```python
import torch

# Sequence length
T = 4

# Dimensions
B = 2
D = 5
H = 4
O = 3

# Input sequence
x = torch.randn(T, B, D)

# Initial hidden
h = torch.zeros(B, H)

# Weights
W_x = torch.randn(H, D)
W_h = torch.randn(H, H)
W_y = torch.randn(O, H)

b = torch.randn(H)
b_y = torch.randn(O)

print("Forward Pass Through Time\n")

for t in range(T):
    h = torch.tanh(x[t] @ W_x.T + h @ W_h.T + b)
    y = h @ W_y.T + b_y

    print(f"Time Step {t+1}")
    print("Hidden shape:", h.shape)
    print("Output shape:", y.shape)
    print("------")
```

# Weight Sharing in RNN

So far we saw:

RNN repeats across time:

$x_1 \rightarrow h_1$  
$x_2 \rightarrow h_2$  
$x_3 \rightarrow h_3$  
$x_4 \rightarrow h_4$

Now the important question:

Do we use different weights at each time step?

Answer:

No.

RNN uses **same weights across all time steps**

---

# Same Weights Across Time

At each time step:

Hidden update:

$$
h_t = \tanh(W_x x_t + W_h h_{t-1} + b)
$$

Notice:

- $W_x$ is same for all time steps
- $W_h$ is same for all time steps
- $W_y$ is same for all time steps

This is called:

# Weight Sharing

---

# Why Weight Sharing?

Because:

Sequential data follows same pattern:

Example:

Sentence:

"I love AI"

"I love ML"

The relationship between words is similar.

So RNN learns:

Same transformation across time.

---

# Parameter Efficiency

Let's compare:

## Without Weight Sharing

If sequence length = 10

Then:

10 different networks required

Parameters:

$$
10 \times (W_x + W_h + W_y)
$$

Huge number of parameters

---

# With Weight Sharing

Only one set of weights:

$$
W_x, W_h, W_y
$$

Used across all time steps

Result:

- Fewer parameters
- Faster training
- Better generalization

---

# Comparison With ANN

ANN:

Each layer has separate weights

$$
W_1, W_2, W_3, W_4
$$

RNN:

Same weights reused across time

$$
W_x, W_h, W_y
$$

---

# Why This is Powerful

Because:

- Sequence length can vary
- Model works for any sequence length
- Efficient learning

This is one of the most important ideas in RNN.

```python
# Example parameter comparison

input_size = 10
hidden_size = 20
output_size = 5
sequence_length = 10

# Without weight sharing
params_without = sequence_length * (
    input_size*hidden_size +
    hidden_size*hidden_size +
    hidden_size*output_size
)

# With weight sharing
params_with = (
    input_size*hidden_size +
    hidden_size*hidden_size +
    hidden_size*output_size
)

print("Parameters without weight sharing:", params_without)
print("Parameters with weight sharing:", params_with)
```

# Backpropagation Through Time (BPTT)

So far we understand:

- RNN forward pass
- Hidden state flow
- Weight sharing

Now the key question:

# How Does RNN Learn?

Like ANN, RNN also uses:

- Loss function
- Gradient descent
- Backpropagation

But there is a difference:

ANN:

Backpropagation across **layers**

RNN:

Backpropagation across **time**

This is called:

# Backpropagation Through Time (BPTT)

---

# Forward Pass Across Time

For sequence:

$x_1, x_2, x_3, x_4$

RNN computes:

$$
h_1 = f(x_1, h_0)
$$

$$
h_2 = f(x_2, h_1)
$$

$$
h_3 = f(x_3, h_2)
$$

$$
h_4 = f(x_4, h_3)
$$

Outputs:

$$
y_1, y_2, y_3, y_4
$$

---

# Loss Across Time

Loss is computed at each time step:

$$
L = L_1 + L_2 + L_3 + L_4
$$

Total loss:

$$
L = \sum_{t=1}^{T} L_t
$$

---

# Backpropagation Through Time

Gradients flow backward:

$$
L_4 \rightarrow h_4 \rightarrow h_3 \rightarrow h_2 \rightarrow h_1
$$

Gradient flows through:

- Time step 4
- Time step 3
- Time step 2
- Time step 1

This is called:

Backpropagation **Through Time**

---

# Why This Happens

Because:

Each hidden state depends on previous hidden state:

$$
h_t = f(x_t, h_{t-1})
$$

So gradient must flow:

$$
\frac{\partial L}{\partial h_4}
\rightarrow
\frac{\partial L}{\partial h_3}
\rightarrow
\frac{\partial L}{\partial h_2}
\rightarrow
\frac{\partial L}{\partial h_1}
$$

This is sequential gradient flow.

---

# Important Observation

Same weights across time:

$$
W_x, W_h, W_y
$$

So gradients from all time steps accumulate:

$$
\frac{\partial L}{\partial W} =
\sum_{t=1}^{T}
\frac{\partial L_t}{\partial W}
$$

This allows RNN to learn:

- Short-term dependencies
- Long-term dependencies

```python
import torch
import torch.nn as nn

# Sequence
T = 5
B = 2
D = 3
H = 4

x = torch.randn(T, B, D)

rnn = nn.RNN(D, H)

output, hidden = rnn(x)

# Dummy target
target = torch.randn(T, B, H)

loss = ((output - target)**2).mean()

print("Loss:", loss.item())

# Backprop
loss.backward()

print("\nGradient for weight:")
print(rnn.weight_ih_l0.grad.shape)
```
