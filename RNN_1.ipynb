# Why CNN is Not Enough for Everything

## Images vs Sequences

So far, we have worked with **Images** using CNNs.

Images have a very specific structure:

- Fixed size
- Spatial relationships
- No time dependency
- All information available at once

Example:

Image → 224 × 224 × 3

CNN processes the **entire image at once**.

---

## But Not All Data Looks Like Images

Many real-world problems involve **Sequential Data**:

Examples:

- Text  
- Speech  
- Time Series  
- Video  
- Sensor Data  

These data types have **order dependency**.

Example:

Sentence:

"I love deep learning"

If we change order:

"Learning deep love I"

Meaning changes completely.

---

## Key Difference

### Images

- Spatial Structure
- Fixed Size
- No Ordering Importance

### Sequences

- Temporal Structure
- Variable Length
- Order Matters

---

## Why CNN Struggles with Sequences

CNN assumes:

- Fixed input size
- No long-term memory
- Local pattern detection

- Variable input length
- Memory of past information
- Context understanding

---

This leads to a fundamental question:

**How do we design neural networks that understand order and memory?**

This is where **Recurrent Neural Networks (RNNs)** come into picture.

# RNN Intuition

We now move from the **need for memory** to the **idea of recurrence**.

In a standard feedforward neural network, each input is processed independently.  
That works well when inputs do not depend on each other.

But in sequential data, the current input often depends on what came before.

Examples:

- In language, the meaning of a word depends on previous words.
- In speech, the current sound depends on earlier sounds.
- In time series, the present value is often related to past values.

So instead of processing each input independently, we want a model that carries forward some information from the past.

This leads to the idea of a **hidden state**.

The hidden state acts like the network's memory.

At time step $t$:

- the network receives the current input $x_t$
- it also uses the previous hidden state $h_{t-1}$
- then it computes the new hidden state $h_t$

The basic RNN update is:

$$
h_t = \tanh(W_x x_t + W_h h_{t-1} + b)
$$

where:

- $x_t$ = current input
- $h_{t-1}$ = previous hidden state
- $h_t$ = new hidden state
- $W_x$ = weight matrix for input
- $W_h$ = weight matrix for previous hidden state
- $b$ = bias

## Why use $\tanh$?

The expression inside the brackets is linear.  
If we keep only linear operations, the model remains too limited.

The $\tanh$ function introduces non-linearity and also keeps the hidden state values bounded between $-1$ and $1$.

This helps in:

- learning more complex relationships
- keeping the hidden state controlled
- preventing values from growing too large during repeated updates

## Main Idea

RNN learns by repeatedly applying:

$$
\text{current input} + \text{previous memory} \rightarrow \text{new memory}
$$

This repeated flow of information across time is what makes the network **recurrent**.

```python
import numpy as np
import matplotlib.pyplot as plt

words = ["I", "love", "deep", "learning"]
time_steps = np.arange(1, len(words) + 1)

plt.figure(figsize=(10, 3))
plt.plot(time_steps, np.ones_like(time_steps), marker='o')

for i, word in enumerate(words):
    plt.text(time_steps[i], 1.05, word, ha='center')
    if i < len(words) - 1:
        plt.arrow(time_steps[i] + 0.1, 1, 0.7, 0,
                  head_width=0.02, head_length=0.08, length_includes_head=True)

plt.yticks([])
plt.xticks(time_steps)
plt.xlabel("Time Step")
plt.title("RNN Intuition: Information Flows Forward Through Hidden State")
plt.ylim(0.9, 1.12)
plt.show()
```

We usually group them into three common task types.

## 1. Many-to-One

A full sequence is given as input, and one output is produced.

Example:

A movie review sentence $\rightarrow$ Positive / Negative

Here the model reads the whole sequence and then makes one final decision.

This is common in:

- sentiment analysis
- spam detection

## 2. One-to-Many

One input is given, and a sequence is generated as output.

Example:

An image $\rightarrow$ a caption sentence

This is common in:

- image captioning
- music generation
- text generation from a starting signal

---

## 3. Many-to-Many

A sequence is mapped to another sequence.

Example:

English sentence $\rightarrow$ Hindi sentence

This is common in:

- language translation
- speech recognition
- subtitle generation
- sequence labeling

---

# Why Simple RNNs Struggle

So far, RNN looks ideal for sequences.

It can:

- process ordered data
- carry memory through hidden states
- work on variable-length inputs

But simple RNNs have an important weakness.

They struggle when useful information must be remembered for a long time.

## Long-Term Dependency Problem

Consider this sentence:

"The book that I bought in Paris last year was written in French, so I could finally understand the ending because I know ______."

To fill the blank correctly, the model may need to remember:

- "Paris"
- "French"

These important clues appeared much earlier in the sequence.

In practice, simple RNN often fails to preserve such information over many time steps.

## Why This Happens

At each time step, the hidden state is updated again and again:

$$
h_t = \tanh(W_x x_t + W_h h_{t-1} + b)
$$

During training, gradients must also flow backward through many time steps.

When this happens repeatedly, gradients can become very small.

This is called the **vanishing gradient problem**.

If gradients become too small:

- early time steps learn very slowly
- distant past information is not preserved well
- long-term dependencies are hard to learn

Sometimes gradients can also become very large.

This is called the **exploding gradient problem**.

## Practical Consequence

Simple RNN works reasonably for short dependencies, but becomes weak when:

- sequences are long
- important context appears far in the past
- memory must be preserved over many steps

## Motivation for Better Models

Because of this limitation, improved recurrent models were introduced:

- LSTM
- GRU

These models are designed to control memory more effectively.

**How do we design a recurrent model that decides what to remember and what to forget?**
